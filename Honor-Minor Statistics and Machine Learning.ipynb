#### WELL-POSED LEARNING PROBLEMS

A computer program is said to learn from experience E with respect
to some class of tasks T and performance measure P, if its performance at tasks in
T, as measured by P, improves with experience E.

For example, a computer program that learns to play checkers might improve
its performance as measured by its ability to win at the class of tasks involving
playing checkers games, through experience obtained by playing games against
itself.

A checkers learning problem:
    <ol>
    <li> Task T: playing checkers</li>
    <li> Performance measure P: percent of games won against opponents</li>
    <li> Training experience E: playing practice games against itself</li>
    </ol>

A handwriting recognition learning problem:
   <ol>
    <li> Task T: recognizing and classifying handwritten words within images</li>
    <li> Performance measure P: percent of words correctly classified</li>
    </ol>

#### DESIGNING A LEARNING SYSTEM

<ol>
    <li> Type of training experience you want to give</li>
    <li> Degree to which learner controls the sequence of training examples</li>
    <li> How well it represent the distribution of learning examples over which final performance of system depends</li>
    </ol>

#### CHOOSING THE TARGET FUNCTION

This step is about the, what type of the knowledge to be learned by the system

Cosider the checker game program where function generates the set of legel moves

Here program need to learn to chose the best move among the all legel moves

Best part of this is how our learning function choses the best move, on which winning depends

$$
ChoseMove:B\rightarrow M
$$    

Function $ChoseMove$ gets board states $B$ as input and produces $M$ as possible Move

Above function will be the choice for target function but it does not deliver the learning ability from the moves. So we have another function 

$$
V:B \rightarrow \mathbf{R}
$$

Here V maps legal states on the board B to the real number $\mathbf{R}$

#### CHOOSING A REPRESENTATION OF THE TARGET FUNCTION

Let us see
<ol>
    <li> $x_1$ be the number of black pieces on the board</li>
    <li> $x_2$ be the number of red pieces on the board</li>
    <li> $x_3$ be the number of black kings on the board</li>
    <li> $x_4$ be the number of red kings on the board</li>
    <li> $x_5$ be the number of black pieces threatened by red</li>
    <li> $x_6$ be the number of red pieces threatened by Black</li>
    </ol>

Thus, the learning program will become

$$
\hat{V}=w_0+w_1x_1+w_2x_2+w_3x_3+w_4x_4+w_5x_5+w_6x_6
$$

Where $w_0$ through $w_6$ are the numerical coefficients or weights to be chosen by the learning algorithm

## Gradient Descent and Cost Function

```python
import numpy as np
import pandas as pd
import seaborn as snb
import matplotlib.pyplot as plt
```

```python
a=np.array([])
b=np.array([])
c=np.array([])
```

```python
def Gradient_Descent(x,y):
    learning_rate=0.0001
    y_pred=0
    m_current=b_current=0
    n_iteration=1500
    n=len(x)
    
    for i in range(n_iteration):
        y_pred=m_current*x+b_current
        a=y_pred
        cost=(1/n)*sum([val**2 for val in (y-y_pred)])
        deri_m=-(2/n)*sum(x*(y-y_pred))
        deri_b=-(2/n)*sum(y-y_pred)
        m_current=m_current-learning_rate*deri_m
        b=m_current
        b_current=b_current-learning_rate*deri_b
        c=b_current
        print("m{}, b{}, cost {}, iteration {}".format(m_current, b_current,cost, i))
    fig = plt.figure()
    ax = fig.add_subplot(111, projection = '3d')
    ax.set_xlabel("Slope")
    ax.set_ylabel("Intercept")
    ax.set_zlabel("Cost")
    ax.scatter(a,b,c)
    #ax = fig.gca(projection='3d')
    #ax.plot_trisurf(a, b, c, cmap=plt.cm.Spectral, linewidth=0.2)
    plt.show()     
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([5,6,9,11,17,19])
Gradient_Descent(x,y)
```
